```python
import pandas as pd
import seaborn as sbn
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
```

```python
plt.rcParams['figure.dpi'] = 150
```

```python
df = pd.read_csv("../data/raw/train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df['newCabin'] = df['Cabin'].str.split(expand=True).apply(lambda s: s.str.slice(stop=1))[0]
```

```python
mosaic(df, ['Survived', 'Embarked'], title='basic dictionary')
plt.show()
```

```python
sbn.violinplot(data=df, y="Age", x="Pclass", hue="Survived", split=True)
```

```python
sbn.violinplot(data=df, y="Age", x="newCabin", hue="Survived", split=False, order='ABCDEFGT')
```

```python
sbn.violinplot(data=df, y="Age", x="Sex", hue="Survived", split=True)
```
